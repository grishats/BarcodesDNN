```python
def evaluate(borders, device='cuda:0'):
    thetas = cut(borders, threshold=95)
    
    dataset='SVHN'
    model_params = {'regularization': 1e-5}

    L = ComputeLoss(model=ResNet, model_params=model_params, dataset=dataset,
                    part='train', bs=150, device=device)
    
    loss_thetas = np.vstack([L.evaluate_thetas(fraction, dataset_ratio=1/8, mode=1/20).reshape(-1, 1) for fraction in thetas])
    
    losses_between = []
    for index in range(math.ceil(len(borders)/2)): #Pick 2 path at max each time
        loss_between = L.evalutae_between_thetas(borders[index * 2: (index + 1) * 2], alpha=3, dataset_ratio=1/8, mode=1/20,
                                                 plot=False)
        losses_between.append(loss_between)
    losses_between = np.vstack(losses_between)
    
    return {'loss_in_thetas' : loss_thetas,
            'losses_between_thetas' : losses_between}
```

```python
sharp_bar, wide_bar = np.load('sharp28.npy', allow_pickle=True).tolist(),\
                      np.load('wide28.npy', allow_pickle=True).tolist()
```

```python
from itertools import combinations

def calculate_barcode(logs, part = 0, N_mins = 6):
        
    pairs = list(combinations(range(N_mins), 2))
    path = []
    
    for i in range(N_mins):
        for j in range(N_mins):
            if j > i:
                path.append([j, i])

    cache = []
    for index, sector in enumerate(path):
        if part in sector:
            cache.append(index)
    
    pottential_barcodes = []
    for index in cache:
        
        if pairs[index][0] == part:
            leading_minimum = logs[index][0]
            other_minimum = logs[index][-1]
            
        else:
            leading_minimum = logs[index][-1]
            other_minimum = logs[index][0]
                        
        if leading_minimum > other_minimum:
            pottential_barcodes.append([leading_minimum, other_minimum, np.max(logs[index])])
            
    if len(pottential_barcodes) == 0:
        pottential_barcodes = leading_minimum
    
    pottential_barcodes = np.array(pottential_barcodes)    
    
    if pottential_barcodes.size == 1:
        print(pottential_barcodes.item())
        out = {'min': pottential_barcodes.item(), 'saddle': None} 
    
    else:
        
        index = np.argmin(pottential_barcodes, axis=0)[-1]
        
        out = {'min': pottential_barcodes[index][0], 'saddle': pottential_barcodes[index][-1], 'pair': index, 'origin': part}
    
    return out
```

```python
N_minimim = 8
df = []
pairs = list(combinations(range(N_minimim), 2))

for index, line in zip(pairs, wide_bar):
    home_idx, destination_idx = index
    
    home, destination = line[0], line[-1]
    saddle = sorted(line)
    
    valid = destination < home
    
    df.append([home_idx, destination_idx, home, destination, saddle, valid])
        
df = pd.DataFrame(df, columns = ['home_idx', 'destination_idx', 'home', 'destination', 'saddle', 'valid'])

reversed_df = pd.concat([df['destination_idx'], df['home_idx'], df['destination'], df['home'], df['saddle'], ~df['valid']],
                         axis=1)

reversed_df.columns = df.columns

DF = pd.concat([df, reversed_df])
DF.index = range(len(DF))

lowest_min = []

for home_idx in DF['home_idx'].unique():
    if DF[DF['home_idx'] == home_idx]['valid'].sum()==0:
        lowest_min.append(home_idx)
        
DF['lowest_minimum'] = DF['home_idx'] == lowest_min[0]

barcodes = []

for home_idx in DF['home_idx'].unique():
    if home_idx not in lowest_min:
        group = DF[(DF['home_idx'] == home_idx) & (DF['valid'] == True)]    
        best_path = group[group['saddle'] == group['saddle'].min()]
    
        barcodes.append([best_path['home_idx'].item(), best_path['destination_idx'].item(), best_path['saddle'].item()])
        
barcodes_df_1 = pd.DataFrame(barcodes, columns=['home_idx', 'destination_idx', 'saddle'])
barcodes_df_1.loc[N_minimim-1] = [lowest_min[0], np.nan, np.nan]
```

```python
N_minimim = 8
df = []
pairs = list(combinations(range(N_minimim), 2))

for index, line in zip(pairs, sharp_bar):
    home_idx, destination_idx = index
    
    home, destination = line[0], line[-1]
    saddle = sorted(line)
    
    valid = destination < home
    
    df.append([home_idx, destination_idx, home, destination, saddle, valid])
        
df = pd.DataFrame(df, columns = ['home_idx', 'destination_idx', 'home', 'destination', 'saddle', 'valid'])

reversed_df = pd.concat([df['destination_idx'], df['home_idx'], df['destination'], df['home'], df['saddle'], ~df['valid']],
                         axis=1)

reversed_df.columns = df.columns

DF = pd.concat([df, reversed_df])
DF.index = range(len(DF))

lowest_min = []

for home_idx in DF['home_idx'].unique():
    if DF[DF['home_idx'] == home_idx]['valid'].sum()==0:
        lowest_min.append(home_idx)
        
DF['lowest_minimum'] = DF['home_idx'] == lowest_min[0]

barcodes = []

for home_idx in DF['home_idx'].unique():
    if home_idx not in lowest_min:
        group = DF[(DF['home_idx'] == home_idx) & (DF['valid'] == True)]    
        best_path = group[group['saddle'] == group['saddle'].min()]
    
        barcodes.append([best_path['home_idx'].item(), best_path['destination_idx'].item(), best_path['saddle'].item()])
        
barcodes_df_0 = pd.DataFrame(barcodes, columns=['home_idx', 'destination_idx', 'saddle'])
barcodes_df_0.loc[N_minimim-1] = [lowest_min[0], np.nan, np.nan]
```

```python
barcodes_df_0.sort_values('saddle', inplace=True)
barcodes_df_1.sort_values('saddle', inplace=True)
res = pd.concat([barcodes_df_0, barcodes_df_1])
```

```python
Number_of_minimums = 8

# [of type 1], [of type 2] both of len 28
trajectories_train = [sharp_bar, wide_bar]


barcodes = []
for index, depth in enumerate(['type1', 'type2']):
    
    depth_barcodes = []
    for minimum in range(Number_of_minimums):
        depth_barcodes.append(calculate_barcode(trajectories_train[index], part = minimum, N_mins=Number_of_minimums))
        
    group = pd.DataFrame(depth_barcodes).sort_values('saddle')
    group['group'] = index
    
    barcodes.append(group)
    
barcodes = pd.concat(barcodes)
barcodes.index = range(len(barcodes))
barcodes['saddle'] = res['saddle'].values

fig, ax = plt.subplots(figsize=(15,10))

prev_group = 0

ax.set_ylim([-0.005, 0.25])

for idx, line in barcodes.iterrows():
    if not np.isnan(line['saddle']):
        ax.plot([idx, idx], [line['min'], line['saddle']], c='gray', linestyle='--', alpha = 0.3)
        ax.scatter([idx], line['saddle'], marker="v", c='mediumblue', s=80)
        ax.scatter([idx], line['min'], marker="*", c='red', s=80)
           
    else:
        ax.scatter([idx], line['min'], marker="*", c='red')
        ax.plot([idx, idx], [line['min'], 2], c='mediumblue', linestyle='-.', alpha=0.2)
    
    if prev_group!=line['group']:
        ax.axvline(idx-0.5, c='r', linewidth = 5, alpha = 0.2)

    prev_group = line['group']


# ax.set_title('Experiment 2, ResNet9', size=20)

ax.scatter([idx], line['saddle'], marker="v", c='mediumblue', label = 'Loss at saddle')
ax.scatter([idx], line['min'], marker="*", c='red', label = 'Loss at minimum')

ax.legend(loc=2, prop={'size': 15})
ax.set_ylabel('Cross-entropy', size = 20)

ax.tick_params(
    axis='x',          # changes apply to the x-axis
    which='both',      # both major and minor ticks are affected
    bottom=False,      # ticks along the bottom edge are off
    top=False,         # ticks along the top edge are off
    labelbottom=True)

ax.set_xticks([3, 11.5])
ax.set_xticklabels(['Small LR', 'Annealing of LR'], fontsize=25)
ax.set_xlim([-0.5, 15.5])

plt.savefig('Experiment2_ResNet9_Barcodes_SVHN.png')
plt.show()
```
